````python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io, transform
from PIL import Image
import matplotlib.pylab as plt
````

````python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]
````

````python
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
````

````python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("Resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)
````

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (lebar): 
  image_3[int((tinggi) / 2), y] = [255,255,255]


final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````

4. Perhatikan potongan kode program berikut:
```
for y in range (lebar): 
  image_3[int((tinggi) / 2), y] = [255,255,255]
  ```
Apakah kegunaan kode [255,255,255]? Jelaskan!

> Kegunaan kode tersebut adalah untuk mengganti 3 layer dari gambar menjadi 255,255,255. Hal itu bermaksud bahwa layer R, G dan B akan memiliki nilai 255, dimana nilai 255, 255, 255 akan menghasilkan warna putih. Jadi, kode tersebut adalah untuk mengganti warna pixel dari warna asli menjadi warna putih.

5. Jelaskan keterkaitan antara pixel dan juga resolusi gambar yang tinggi ataupun
rendah!

> Resolusi gambar adalah jumlah pixel yang digunakan untuk membentuk sebuah gambar, dalam hal ini resolusi mempengaruhi kualitas gambar. Gambar dengan resolusi tinggi memiliki jumlah pixel yang lebih banyak begitupun sebaliknya. Sehingga, gambar dengan resolusi tinggi akan memiliki tampilan yang lebih jelas dan detail dibandingkan dengan gambar dengan resolusi rendah.  





````python
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range (tinggi): 
  image_5[x, int((lebar) / 2)] = [255,255,255]


final_frame_2 = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame_2)
````

````python
image_7 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range(tinggi):
  image_7[x,x] = [255,255,255]
  image_7[x, tinggi-x-1] = [255,255,255]

final_frame_3 = cv.hconcat((image_2, image_7))
cv2_imshow(final_frame_3)
````

**Note**
- Dikarenakan gambar tidak memiliki square shape, maka garis yang dihasilkan dengan menggunakan perulangan adalah seperti di atas. 
- Namun, ada cara lain untuk menggambar garis secara diagonal dari ujung ke ujung dengan tepat menggunakan bantuan dari cv.line, sebagai berikut:

````python
image_x = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for i in range(tinggi):
  image_x[i, int(i * (lebar/tinggi))] = [255,255,255]
  image_x[i, lebar-1-(int(i*(lebar/tinggi)))] = [255,255,255]


final_frame_x = cv.hconcat((image_2, image_x))
cv2_imshow(final_frame_x)
````

````python
image_8 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

image_8 = cv.line(image_8,(0,0),(lebar,tinggi),[255,255,255],1)
image_8 = cv.line(image_8,(0,tinggi),(lebar,0),[255,255,255],1)


final_frame_6 = cv.hconcat((image_2, image_8))
cv2_imshow(final_frame_6)
````

````python
image_9 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(100, 200): 
  image_9[int((tinggi) / 2), y] = [255,255,255]


final_frame_4 = cv.hconcat((image_2, image_9))
cv2_imshow(final_frame_4)
````
